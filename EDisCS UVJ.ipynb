```python
import pyfits 
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
```

```python
f = open('/Users/jennifercooper/Documents/Gal_Ev/EDisCS/UVJ-HaLPDmatch.txt', 'r')
lines = f.readlines()[1:]
f.close()

#create arrays 
f_Ha    = [] 
rf_J    = []
rf_U    = []
rf_V    = []
rf_VJ   = []
rf_UV   = []

#pull array column 
for line in lines: 
    a = line.split()
    f_Ha.append(float(a[0]))
    rf_J.append(float(a[1]))
    rf_U.append(float(a[2]))
    rf_V.append(float(a[3]))
    rf_VJ.append(float(a[4]))
    rf_UV.append(float(a[5]))
    
f_Ha = np.array(f_Ha)
rf_J = np.array(rf_J)
rf_U = np.array(rf_U)
rf_V = np.array(rf_V)
rf_VJ = np.array(rf_VJ)
rf_UV = np.array(rf_UV)
```

```python
plt.scatter(rf_VJ,rf_UV,c=f_Ha,alpha=0.2)
plt.xlim(-4,4)
plt.ylim(-2,4)
plt.xlabel('V-J')
plt.ylabel('U-V')
plt.title('All pointings matched with LPD')
cbar = plt.colorbar()
cbar.set_label("H alpha")
plt.show()
```

```python
plt.scatter(rf_VJ[np.where(f_Ha>0)],rf_UV[np.where(f_Ha>0)],c=f_Ha[np.where(f_Ha>0)],alpha=0.2)
plt.xlim(-4,4)
plt.ylim(-2,4)
plt.xlabel('V-J')
plt.ylabel('U-V')
plt.title('All pointings matched with LPD Halpha>0')
cbar = plt.colorbar()
cbar.set_label("H alpha")
plt.show()
```

```python
plt.scatter(rf_V[np.where(f_Ha>0)],rf_UV[np.where(f_Ha>0)],c=f_Ha[np.where(f_Ha>0)],alpha=0.2)
plt.xlim(0,50)
plt.ylim(-1,3)
plt.xlabel('V')
plt.ylabel('U-V')
plt.title('All pointings matched with LPD Halpha>0')
cbar = plt.colorbar()
cbar.set_label("H alpha")
plt.show()
```

```python
plt.scatter(rf_VJ[np.where(f_Ha>0)],rf_UV[np.where(f_Ha>0)],c='red',alpha=0.2)
plt.scatter(rf_VJ[np.where(f_Ha<1E-21)],rf_UV[np.where(f_Ha<1E-21)],c='blue',alpha=0.2)
plt.xlim(-4,3)
plt.ylim(-1,3)
plt.xlabel('V-J')
plt.ylabel('U-V')
plt.title('All pointings matched with LPD Halpha>0')
red_patch = mpatches.Patch(color='purple', label='H alpha Detection', alpha = 0.5)
blue_patch = mpatches.Patch(color='blue', label='No detection', alpha = 0.5)
plt.legend(handles=[red_patch,blue_patch])
    
plt.show()
```
